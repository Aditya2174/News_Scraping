```python
#import necessary libraries here
import requests
from bs4 import BeautifulSoup
```

```python
#web scraping logic goes here, make sure to add proper comments for better readability of your code
web=requests.get("https://www.moneycontrol.com/news/business/economy/").content
```

```python
soup=BeautifulSoup(web,'lxml')
```

```python
headlines = []

for h2_tag in soup.find_all("h2"):                # First, iterate over all h2 tags in the HTML content
    a_tag = h2_tag.find('a')                      # Then, find the first 'a' tag within the current h2 tag
    if a_tag and 'title' in a_tag.attrs:          # Check if the 'a' tag exists and has a 'title' attribute
        headlines.append(a_tag.attrs['title'])    # Append the value of the 'title' attribute to the list

# Print the list
headlines

```
